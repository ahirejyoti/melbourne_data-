```
from scipy.stats import skew
#skewness check
```

```
df_cat=df.select_dtypes(object)
df_num=df.select_dtypes(["int64","float64"])
df_cat.head()
```

```
for col in df_num:
    print(col,":", skew(df_num[col]))
    sns.distplot(df_num[col])
    plt.show()
    print("--------")
```

```
plt.subplots(figsize=(20,8))

sns.heatmap(df_num.corr(), annot=True)

plt.show()
```

```
from sklearn.preprocessing import LabelEncoder
```

```
for col in df_cat:
      le=LabelEncoder()
      df_cat[col]=le.fit_transform(df_cat[col])
      df_new=pd.concat([df_num,df_cat], axis=1)
      
```

```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```
x=df_new.drop(col, axis=1)
y=df_new[col]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```
lr=LinearRegression()
lr.fit(x_train, y_train)
print("""Slope ={}
intercept = {}
Bias ={}
variance = {}"""
.format(lr.intercept_,lr.coef_,lr.score(x_train,y_train),lr.score(x_test,y_test )))
```

```
# Bias & variance are almost equal means this is good fit graph.
# no need of crossvalidation
```

```
from sklearn.linear_model import Lasso,Ridge
```

```
r=Ridge(0.005)
r.fit(x_train,y_train)
cols=x_train.columns
coef = pd.Series(lr.coef_,cols).sort_values()
coef.plot(kind="bar",title="Model coefficients/slopes")
```

```
coef
```

```
for i in range (50):
    r=Ridge(alpha=i)
    r.fit(x_train,y_train)
    print(i,"=",r.score(x_test,y_test))
```

```
r=Lasso(2)
r.fit(x_train,y_train)
cols=x_train.columns
coef = pd.Series(lr.coef_,cols).sort_values()
coef.plot(kind="bar",title="Model coefficients/slopes")
plt.show()
```

```
for i in range (2,2,2):
    r=Lasso(alpha=i)
    r.fit(x_train,y_train)
    print(i, "=",r.score(x_test,y_test))
```
